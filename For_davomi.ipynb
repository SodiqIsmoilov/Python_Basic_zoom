```python
# 1.  n natural soni berilgan. 1 dan n gacha bo’lgan sonlarning yig'indisini 
# hisoblovchi dastur tuzing.
n = int(input('n = '))
s = 0
for i in range(n+1):
    s+=i
print(s)
```

```python
# 3. Kiritilgan sonning tub yoki tub emasligini aniqlovchi dastur tuzing. 
n = int(input('n = '))
s = 0
for i in range(1,n+1):
    if not(n%i):
        s+=1
if s==2:
    print('Tub')
else:
    print('Murakkab')
```

```python
# 4. 1 dan n gacha bo’lgan sonlar oralig’idagi tub sonlarni va ularning sonini 
# ekranga chiqaruvchi dastur tuzing 
n = int(input('n = '))
count = 0
for i in range(1,n+1):
    s = 0
    for j in range(1,i+1):
        if i%j==0:
            s+=1
    if s==2:
        print(i, end=' ')
        count+=1
print()
print(f'1 dan {n} gacha bo\'lgan tub sonlarning umumiy soni:',count)
```

```python
# 5. n soni berilgan. n faktorialni hisoblovchi dastur tuzing ( n!=1*2*3*…*n ).
n = int(input('n = '))
f = 1
for i in range(1,n+1):
    f*=i
print(f'{n}!={f}')
```

```python
# 6. n soni berilgan. n gacha bo’lgan tub sonlarning oxirgisini aniqlovchi dastur tuzing. 
n = int(input('n = '))
loop = True
for i in range(n,0,-1):
    s = 0
    if loop:
        for j in range(1,j+1):
            if i%j==0:
                s+=1
        if s==2:
            loop=False
            print(f'Oxirgi tub son {i}')        
```
