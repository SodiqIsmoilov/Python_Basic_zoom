```python
# 29. n musbat soni berilgan. n-darajali Paskal uchburchagini piramida ko’rinishida   (funksiya bilan qilingan)
# ekranga chiqaruvchi dastur tuzing.
# Kiritish: 5
# Chiqarish:                                  
#          1                                                            
#        1   1                                                          
#      1   2   1                                                        
#    1   3   3   1                                                      
#  1   4   6   4   1
def factarial(x):
    f = 1
    for i in range(1,x+1):
        f*=i   
    return f
def Cnk(n,k):
    return factarial(n)/(factarial(k)*factarial(n-k))

n = int(input('n='))
for i in range(n):
    print('\n',end=' '*(n-i))
    for j in range(i+1):
        print(int(Cnk(i,j)),end=' ')
```
