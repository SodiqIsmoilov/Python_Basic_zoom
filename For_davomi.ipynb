```python
# 6.                    n soni berilgan. n gacha bo’lgan tub sonlarning oxirgisini aniqlovchi dastur tuzing. 
n = int(input('n = '))
loop = True
for i in range(n,0,-1):
    s = 0
    if loop:
        for j in range(1,j+1):
            if i%j==0:
                s+=1
        if s==2:
            loop=False
            print(f'Oxirgi tub son {i}')        
```

```python
# 7.                   Ikki sonning eng katta umumiy bo'linuvchisini EKUB(a,b) aniqlovchi dastur tuzing.
a = int(input('a ='))
b = int(input('b ='))
ekub = 0
i = 1
while (i<a or i<b):
    if a%i==0 and b%i==0:
        ekub=i
    i+=1
print(ekub)
```

```python
#                    Uchta sonning eng katta umumiy bo'linuvchisini EKUB(a,b,c) aniqlovchi dastur tuzing.
a = int(input('a ='))
b = int(input('b ='))
c = int(input('c ='))
ekub = 0
if a>b>c or b>a>c:
    a = c
elif a>c>b or c>a>b:
    a = b
for i in range(1,a+1):
    if a%i==0 and b%i==0 and c%i==0:
        ekub = i
if ekub==1:
    print('O`zaro tub')
else:
    print(ekub)
```

```python
# 8.               Berilgan sonning raqamlari yig'indisini aniqlovchi dastur tuzing.    (for)
n = int(input('n ='))
s = 0
for i in range(1,n):
    xona = n%10**i
    xona //= 10**(i-1)
    s+=xona
    if xona==0:
        break
print(s)
```

```python
# 8.              Berilgan sonning raqamlari yig'indisini aniqlovchi dastur tuzing.    (while)
n = int(input('n ='))
loop = True
i=1
s=0
while loop:
    xona = n%10**i
    xona //= 10**(i-1)
    s+=xona
    if xona == 0:
        loop=False
    i+=1
print(s)
```

```python
# 29. n musbat soni berilgan. n-darajali Paskal uchburchagini piramida ko’rinishida 
# ekranga chiqaruvchi dastur tuzing.
n = int(input('n='))
def factarial(x):
    f = 1
    for i in range(1,x+1):
        f*=i
    return f
def Cnk(n,k):
    return factarial(n) / (factarial(k)*factarial(n-k))
tab = n
for i in range(n):
    print(end=' '*tab)
    for j in range(i+1):
        print(int(Cnk(i,j)), end=' ')
    print()
    tab-=1
```
