# Tub sonlar

```python
n = int(input('n='))
count = 0
for i in range(2,n+1,):
    s=0
    for j in range(1,i+1):
        count+=1
        if i%j==0:
            s+=1
    if s==2:
        print(i,end=' ')
print('\nSikl=',count, end=' ')
```

```python
from math import sqrt
n = int(input('n='))
count = 0
print(2,end=' ')
for i in range(3,n+1,2):
    s=0
    for j in range(3,int(sqrt(i))+1,2):
        count+=1
        if i%j==0:
            s+=1
            break
    if s==0:
        print(i,end=' ')
print('\nSikl',count, end=' ')
```

# Sanoq sistemasi Kalkulator

```python
n = int(input('n = '))
berilgan_sis = int(input('qaysi sanoq sistemasida kiritdingiz = '))
topish_sis = int(input('qaysi sanoq sistemasiga o`tkazish kerak = '))
#   berilgan_sistema --->   10    ---> topish_sitema

# ixtiyoriydan 10 lik sanoq sistemasiga o`tkazish
s=0
i=0
while n!=0:
    s+=(n%10)*berilgan_sis**i
    n//=10
    i+=1
# 10 likdan so`ralgan sistemaga o`tish
j=0
javob=0
while s>0:
    javob+=((s%topish_sis)*10**j)    
    s//=topish_sis    
    j+=1
print(javob)
```

# Do'st sonlar

```python
#version 1.0
n = int(input('n='))
count = 0
r = 0
for i in range(2,n+1):
    s1 = 0
    s2 = 0
    for j in range(1,i):
        if i%j==0:
            s1+=j    
    for k in range(1,s1):
        if s1%k==0:
            count+=1
            s2+=k
    if i==s1==s2 or i==r:
        continue
    if i==s2:
        print(i,'s1=',s1,'s2=',s2,end='\n')
        r = s1
print('count=',count,end='')
```

```python
#version 2.0
n = int(input('n='))
count = 0
for i in range(2,n+1,2):
    s1 = 1
    s2 = 1    
    for j in range(2,int(i/2)+1):
        if i%j !=0:
            continue
        s1+=j        
    
    if s1<i or s1==1:        
        continue
    
    for k in range(2,int(s1/2)+1):
        if s1%k!=0:
            continue
        s2+=k
    
    if i==s1==s2 or i==r:
        continue
    
    if i==s2:
        print(i,'s1=',s1,'s2=',s2,end='\n')
        r = s1
print('count=',count,end='')
```
