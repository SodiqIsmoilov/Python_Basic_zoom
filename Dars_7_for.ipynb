```python
#for 27
x = int(input('x soni :'))
n = int(input('n soni :'))
t = 1   # toqlar uchun
j = 1   #juftlar uchun
s = x   #summa
for i in range(1,n+1):
    t *= (2*i-1)
    j *= (2*i)
    s += (t/j) * x**(2*i+1) / (2*i+1)
    
print(s, end=' ')
```
