```python
#Array 96
n = int(input('n='))
A = [randint(1,10) for i in range(n)]
print(A)
i = 0
while i<len(A):
    if A.count(A[i])>1:
        A = A[::-1]
        while A.count(A[i])!=1:
            A.remove(A[i])
        A = A[::-1]
    i+=1
print(A)
```
