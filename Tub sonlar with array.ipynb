```python
from random import randint
from math import sqrt
```

```python
n = int(input('n='))
tub = [3,5,7]
count = 0
for i in range(11,n+1,2):    
    j = 0
    while tub[j]<int(sqrt(i))+1:
        count+=1
        if i%tub[j]==0:
            break
        j+=1
    else:
        tub.append(i)
print(*tub)
print('Count=',count)
```

```python
n = int(input('n='))
tub = [3,5,7,11]
count = 0
for i in range(13,n+1,2):
    loop = int(sqrt(i))+1
    j = 0
    while tub[j]<=loop:
        count+=1
        if i%tub[j]==0:
            break
        j+=1
    else:
        tub.append(i)

print(*tub)
print('Count =', count)
```

```python
from datetime import datetime
n = int(input('n='))
start_time = datetime.now()
i = 0
tub=[2,3]
temp_tub = [i for i in range(3,n+1,2) if i%3!=0]
count = 0
while tub[i]<int(sqrt(n))+1:
    count+=1
    elem = temp_tub[0]
    temp_tub1 = temp_tub.copy()
    temp_tub = [i for i in temp_tub1 if i%elem!=0]
    tub.append(elem)
    i+=1
tub.extend(temp_tub)
print(*tub)
print('Count=',count, 'Time=',datetime.now() - start_time)
```
