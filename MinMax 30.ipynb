```python

N = int(input('N='))
x =True
i = 1
while i<=N:
    n = int(input('n='))
    if x:
        n_oldin = n
        maxim_number = n_oldin
        javob = N
        temp_smallest = 0
        x = False
    if n == n_oldin:
        temp_smallest += 1
    else:
        if n_oldin > maxim_number:
            javob = temp_smallest
            maxim_number = n_oldin
        elif n_oldin == maxim_number and temp_smallest < javob:
            javob = temp_smallest
        temp_smallest = 1
    n_oldin = n
    i+=1
    

if n > maxim_number:
    javob = temp_smallest 
    maxim_number = n_oldin
elif n == maxim_number and temp_smallest < javob:
    javob = temp_smallest


print(javob)
```
