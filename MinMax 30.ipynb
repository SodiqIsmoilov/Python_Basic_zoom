```python

N = int(input('N='))
x =True
i = 1
while i<=N:
    n = int(input('n='))
    if x:
        n_oldin = n
        maxim_number = n_oldin
        javob = N
        t = 0
        x = False
    if n == n_oldin:
        t += 1
    else:
        if n_oldin > maxim_number:
            javob = t
            maxim_number = n_oldin
        elif n_oldin == maxim_number and t < javob:
            javob = t
        t = 1
    print('t =',t,'javob=', javob)
    n_oldin = n
    i+=1
    

if n > maxim_number:
    javob = t
elif n == maxim_number and t < javob:
    javob = t


print(javob)
```
