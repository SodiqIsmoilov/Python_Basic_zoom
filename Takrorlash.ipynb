```python
#min max 1
N = int(input('N='))
x = True
for i in range(N):
    n = int(input('n='))
    if x:
        minum,maxim = n,n
        x=False
    if maxim<n:
        maxim=n
    if minum>n:
        minum = n
print(maxim,minum)
```
