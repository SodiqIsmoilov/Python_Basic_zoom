```python
#func 4
from math import sqrt
def Triangle(a,b,c):
    P = a+b+c
    p = P/2
    S = sqrt(p*(p-a)*(p-b)*(p-c))
    return P, S
a = float(input('a='))
b = float(input('b='))
c = float(input('c='))
P , S = Triangle(a,b,c)
print('Perimetri=',P,'Yuzi=',S)
```

```python
#func 60
from math import sqrt
def Leng(x1,y1,x2,y2):
    return sqrt((x1-x2)**2+(y1-y2)**2)
def yuza(a,b,c):
    P = (a+b+c)/2
    s = sqrt(P*(P-a)*(P-b)*(P-c))
    return s
def Heights(x1,y1,x2,y2,x3,y3):
    a = Leng(x1,y1,x2,y2)
    b = Leng(x2,y2,x3,y3)
    c = Leng(x3,y3,x1,y1)
    S = yuza(a,b,c)
    ha = 2*S/a; hb = 2*S/b; hc = 2*S/c;
    return ha, hb,hc
x1 = float(input('x1='))
y1 = float(input('y1='))
x2 = float(input('x2='))
y2 = float(input('y2='))
x3 = float(input('x3='))
y3 = float(input('y3='))
print(Heights(x1,y1,x2,y2,x3,y3))    
```

```python
# N ta sonning EKUK topish
def EKUK(a,b):
    a1,b1 = a,b
    while a!=b:
        if a>b:
            a%=b
        else:
            b%=a
        if a==0:
            a=b
        if b==0:
            b=a
    return a1*b1/a
n = int(input('n='))
a = int(input('a='))   
b = int(input('b='))
javob = EKUK(a,b)
i = 0
while n-2>i:
    c = int(input('c ='))
    javob = EKUK(javob,c)
    i+=1
print(javob)
```

```python
# Berilgan sanadan n kun keyingi va oldingi kunni topish
def IsLeapYear(y):
    kabisa = False
    if y%4==0:
        kabisa = True
    if y%100==0 and y%400!=0:
        kabisa = False
    return kabisa

def MonthDays(m,y):
    if m==4 or m==6 or m==9 or m==11:
        return 30
    elif m==2:
        return 28+IsLeapYear(y)
    elif m==1 or m==3 or m==5 or m==7 or m==8 or m==10 or m==12:
        return 31
def NextDate(d,m,y,n):
    while n!=0:
        d+=1
        n-=1
        if d>MonthDays(m,y):
            d = 1
            m = m + 1
        if m==13:
            m = 1
            y = y + 1
    return d,m,y
d = int(input("d = "))
m = int(input("m = "))
y = int(input("y = "))
n = int(input('n = '))
print(NextDate(d,m,y,n))
```
