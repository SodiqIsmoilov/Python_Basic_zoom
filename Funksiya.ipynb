```python
#func 1
def PowerA3(x):
    s = 1
    for i in range(3):
        s *=x
    return s
A = float(input('A='))
B = float(input('B='))
C = float(input('C='))
D = int(input('D='))
E = int(input('E='))
print('A^3 =', PowerA3(A))
print('B^3 =', PowerA3(B))
print('C^3 =', PowerA3(C))
print('D^3 =', PowerA3(D))
print('E^3 =', PowerA3(E))
```

```python
#func 2
def PowerA234(a,n):
    s = 1
    for i in range(n):
        s*=a
    return s
A = float(input('A='))
B = float(input('B='))
C = float(input('C='))
print('A^2 =', PowerA234(A,2),'A^3 =', PowerA234(A,3),'A^4 =', PowerA234(A,4))
print('B^2 =', PowerA234(B,2),'B^3 =', PowerA234(B,3),'B^4 =', PowerA234(B,4))
print('C^2 =', PowerA234(C,2),'C^3 =', PowerA234(C,3),'C^4 =', PowerA234(C,4))
```

```python
#func 3
from math import sqrt
def MEAN(a,b):
    arifmetik = (a+b)/2
    geometrik = sqrt(a*b)
    return arifmetik,geometrik
A = float(input('A='))
B = float(input('B='))
C = float(input('C='))
D = float(input('D='))
print('A va B', MEAN(A,B))
print('A va C', MEAN(A,C))
print('A va D', MEAN(A,D))
```

```python
#func 4
from math import sqrt
def Triangle(a,b,c):
    P = a+b+c
    p = P/2
    S = sqrt(p*(p-a)*(p-b)*(p-c))
    return P, S
a = float(input('a='))
b = float(input('b='))
c = float(input('c='))
P , S = Triangle(a,b,c)
print('Perimetri=',P,'Yuzi=',S)
```

```python
#func 5
def RectPS(x1,y1,x2,y2):
    a = abs(x1-x2)
    b = abs(y1-y2)
    P = 2*(a+b)
    S = a*b
    return P, S
x1 = float(input('x1='))
y1 = float(input('y1='))
x2 = float(input('x2='))
y2 = float(input('y2='))
p , s = RectPS(x1,y1,x2,y2)
print('Perimetri=',p,'Yuzi=',s)
x1 = float(input('x1='))
y1 = float(input('y1='))
x2 = float(input('x2='))
y2 = float(input('y2='))
p , s = RectPS(x1,y1,x2,y2)
print('Perimetri=',p,'Yuzi=',s)
```

```python
#func 6
def DigitCountSum(n):
    s = 0
    count = 0
    while n!=0:
        s +=(n%10)
        n //=10
        count +=1
    return count, s
a = int(input('a='))
count, summ = DigitCountSum(a)
print('Raqamlar soni=', count, 'Raqamlar yi`gindisi=',summ)
b = int(input('b='))
count, summ = DigitCountSum(b)
print('Raqamlar soni=', count, 'Raqamlar yi`gindisi=',summ)
c = int(input('c='))
count, summ = DigitCountSum(c)
print('Raqamlar soni=', count, 'Raqamlar yi`gindisi=',summ)
```

```python
#func 12
def Sortlnc3(a,b,c):
    if a>c:
        a,c = c,a
    if b>c:
        b,c = c,b
    if a>b:
        a,b = b,a
    return a,b,c
a = int(input('a='))
b = int(input('b='))
c = int(input('c='))
print(Sortlnc3(a,b,c))
```
