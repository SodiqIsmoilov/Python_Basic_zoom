```python
#Array 121
n = int(input('n='))
k = int(input('k='))
A = [randint(1,4) for i in range(n)]
print(A)
x = A[0]
k -=1
i = 1
while i<len(A):
    if A[i]!=x:
        x = A[i]
        k-=1
    if k==0:
        print('i=',i)
        while A[i] == x:
            A.insert(i,x)
            i+=2
            if i>len(A)-1:
                break
        break
    i+=1
print(A)
```
