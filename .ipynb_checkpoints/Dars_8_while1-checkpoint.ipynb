```python
#  8
from math import sqrt
n = int(input('n='))
count = 0
for i in range(1,n+1):
    for j in range(1,n+1):        
        for k in range(1,n+1):
#             if i+j<k:
#                 continue
            count+=1
            if k==sqrt(pow(i,2)+pow(j,2)):
                print(i,j,k)
print('count=',count)
```

```python
#while 23   EKUB(a,b)   (oddiy)
a = int(input('a='))
b = int(input('b='))
i = 0
count = 0
while i<a or i<b:
    i+=1
    if a%i != 0 and b%i !=0:
        continue
    if a%i == 0 and b%i ==0:
        ekub=i
    
    count+=1
print(ekub,count)
```

```python
#while 23   EKUB(a,b)   (optimal)
a = int(input('a='))   
b = int(input('b='))
if b>a:
    a,b = b,a
c = 0    # count
while b!=0:
    c+=1
    a%=b
    a,b = b,a
print(a,c)
```
