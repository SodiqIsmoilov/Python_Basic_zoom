```python
#MinMax 1
N = int(input('N='))
i = 0
minum = True

while i<N:
    n=int(input('n='))
    if minum:
        kichik = n
        katta = n
        minum = False
    if n<kichik:
        kichik = n
    if n>katta:
        katta = n    
    i+=1
print(kichik, katta)
```

```python
#MinMax 2
N = int(input('N='))
i = 0
loop = True
while i<N:
    a,b = map(int, input('a va b (a probil b shaklda kiriting)').split(' '))
    if loop:
        s = a*b
        loop = False
    if a*b<s:
        s = a*b
    i+=1
print(s/2)
```
