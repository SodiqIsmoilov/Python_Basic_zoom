```python
#Matrix 15
n = int(input("m = ")) 
matrix = [[0 for _ in range(n)] for _ in range(n)]

elem = 0
satr_boshi = 0; satr_oxiri = n-1
ustun_boshi = 0; ustun_oxir = n-1
while ustun_boshi <= ustun_oxir  and satr_boshi <= satr_oxiri:
    
    for j in range(ustun_boshi,ustun_oxir +1):
        elem += 1
        matrix[satr_boshi][j] = elem
    satr_boshi += 1
    
    for j in range(satr_boshi,satr_oxiri+1):
        elem += 1
        matrix[j][ustun_oxir] = elem
    ustun_oxir  -= 1
    
    for j in range(ustun_oxir ,ustun_boshi-1,-1):
        elem += 1
        matrix[satr_oxiri][j] = elem
    satr_oxiri -= 1
    
    for j in range(satr_oxiri,satr_boshi-1,-1):
        elem += 1
        matrix[j][ustun_boshi] = elem
    ustun_boshi += 1

for i in matrix:
    print(*i)
```

```python
#Matrix 48
m = int(input('m='))
n = int(input('n='))
matrix = [[randint(1,9) for _ in range(n)] for _ in range(m)]
k1 = int(input('k1='))
k2 = int(input('k2='))

for i in matrix:
    print(*i)
print()

if 0<=k1<k2<m:
    for i in range(len(matrix)):
        matrix[i][k1], matrix[i][k2] = matrix[i][k2], matrix[i][k1]
else:
    print('k1 va k2 ni 0<=k1<k2<m oraliqdan kiritng')

for i in matrix:
    print(*i)
print()
```

```python
#Matrix 49
m = int(input('m='))
n = int(input('n='))
matrix = [[randint(1,9) for _ in range(n)] for _ in range(m)]

for i in matrix:
    print(*i)
print()

for i in range(len(matrix)):
    mini = matrix[i].index(min(matrix[i]))
    maxi = matrix[i].index(max(matrix[i]))
    matrix[i][mini], matrix[i][maxi] = matrix[i][maxi], matrix[i][mini]

for i in matrix:
    print(*i)
print()
```

```python
#Matrix 50
m = int(input('m='))
n = int(input('n='))
matrix = [[randint(1,9) for _ in range(n)] for _ in range(m)]

for i in matrix:
    print(*i)
print()

for j in range(len(matrix[0])):
    x = True
    for i in range(len(matrix)):
        if x:
            mini = maxi = matrix[i][j]
            index_mini, index_maxi = i, i
            x = False
        if maxi<matrix[i][j]:
            maxi = matrix[i][j]
            index_maxi = i
        if mini>matrix[i][j]:
            mini = matrix[i][j]
            index_mini = i
    matrix[index_maxi][j], matrix[index_mini][j] = matrix[index_mini][j], matrix[index_maxi][j]

for i in matrix:
    print(*i)
print()
```
