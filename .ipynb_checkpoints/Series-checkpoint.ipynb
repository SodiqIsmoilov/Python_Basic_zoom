```python
#Series 1
s = 0
for _ in range(10):
    n = int(input('n='))
    s+=n
print(s)
```

```python
#Series 2
s = 1
for _ in range(5):
    n = int(input('n='))    
    s *=n
print(s)
```

```python
#Series 3
s = 0
for _ in range(10):
    n = int(input('n='))
    s+=n
else:
    print(s/10)
```

```python
#Series 4
N = int(input('N='))
s = 0
p = 1
for _ in range(N):
    n = int(input('n='))
    s+=n
    p*=n
else:
    print('Yig`indi=',s)
    print('Ko`paytma=',p)
```

```python
#Series 5
N = int(input('N='))
s = 0
for _ in range(N):
    n = float(input('n='))
    s += n//1
else:
    print(s)
```

```python
#Series 6
N = int(input('N='))
s = 0
for _ in range(N):
    n = float(input('n='))
    s += round(n%1,20)                          #aniq chiqishi uchun round dan foydalanilgan
    print('Kasr qismi = ',round(n%1,20))
else:
    print('Kars qismlar yig`indisi=',s)
```

```python
#Series 7
N = int(input('N='))
s = 0
for _ in range(N):
    n = float(input('n='))
    print(round(n))
    s+=round(n)
else:
    print(s)
```
