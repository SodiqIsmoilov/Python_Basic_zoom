```python
#Array 3
a = int(input('a='))
d = int(input('d='))
n = int(input('n='))
arifmetik_prog = [a]
for i in range(2,n+1):
    a += d
    arifmetik_prog.append(a)
print(arifmetik_prog)
```

```python
#Array 5
n = int(input('n='))
f1 = 1
f2 = 1
fibo = []
for i in range(2,n+1):
    fibo.append(f2)    
    f1,f2 = f1+f2, f1
print(fibo)
```

```python
#Array 6
n = int(input('n='))
A = int(input('A='))
B = int(input('B='))
a=[A,B]
for i in range(3,n+1):
    A,B = B,A+B
    a.append(B)
print(a)
```

```python
#Array 8
array8 = [7, 10, 9, 10, 1, 7, 9, 6, 5, 8]
s=0
for i in range(len(array8)):
    if array8[i]%2==1:
        print(array8[i],end=' ')
        s +=1
print('toqlar soni =', s)
```

```python
#Array 9
array9 = [4,5,7,8,6,9]
javob = []
s=0
for i in range(len(array9)):
    if array9[i]%2==0:
        javob.append(array9[i])
        s +=1
print(javob[::-1],end=' ')
print('juftlar soni =', s)
```

```python
#Array 10
array10 = [6, 10, 10, 4, 4, 6, 7, 4, 6, 9]
juftlar = []
toqlar = []
for i in range(len(array10)):
    if array10[i]%2==0:
        juftlar.append(array10[i])
        continue
    toqlar.insert(0,array10[i])
juftlar.extend(toqlar)
print(juftlar)
```

```python
#Array 13 v1.0    shartida xatoligi bor

array13 = [randint(1,10) for _ in range(9)]
print(array13)
i=1
n = len(array13)
while n-i>1:
    print(array13[n-i], end=' ')
    i+=2
```
