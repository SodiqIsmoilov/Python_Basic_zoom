```python
#MinMax - 1
N = int(input('N='))
i = 1
katta = kichik = int(input('n ='))
while i<N:
    n = int(input('n='))
    if n>katta:
        katta = n
    if n<kichik:
        kichik = n
    i+=1
print('Max =',katta,'Min =', kichik)
```

```python
#MinMax - 2
N = int(input('N='))
i = 1
a = float(input('a ='))
b = float(input('b ='))
s = a*b
while i<N:
    a = float(input('a ='))
    b = float(input('b ='))
    if s>a*b:
        s = a*b
    i+=1
print('Eng kichik yuza =', s/2)
```

```python
#MinMax - 3
from math import sqrt
N = int(input('N='))
i = 1
a = float(input('a ='))
b = float(input('b ='))
p = a+b+sqrt(pow(a,2)+pow(b,2))
while i<N:
    a = float(input('a ='))
    b = float(input('b ='))
    if p<a+b+sqrt(pow(a,2)+pow(b,2)):
        p = a+b+sqrt(pow(a,2)+pow(b,2))
    i+=1
print('Eng katta perimetr =', p)
```
